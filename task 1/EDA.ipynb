```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

# Crea una mappa dei colori personalizzata con un colore più arancione e nero
from matplotlib.colors import LinearSegmentedColormap

# Definisci i colori
colors = ["#00FF5987", "#0c0b00"]  # Nero e un arancione più arancione

# Crea la mappa dei colori
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors, N=2)

# Personalizza la dimensione della figura per la leggibilità
plt.figure(figsize=(24, 16))

# Trasponi il DataFrame per invertire gli assi
sns.heatmap(ds.isnull().T, cbar=False, cmap=cmap)

# Aggiungi titolo e etichette per chiarezza
plt.xlabel("Features", fontsize=24)  # Features sull'asse x ora
# Rimuovi i numeri sull'asse delle x
plt.xticks([])
plt.yticks(fontsize=24)

plt.show()
```

```python
# raggruppo per artista e calcolo avg_streams = sum(streams)/count(tracks)
top_artists = (ds.groupby('primary_artist')
               .agg({'streams@1month': 'sum', 'id': 'count'})
               .assign(avg_streams=lambda x: x['streams@1month'] / x['id'])
               .sort_values('avg_streams', ascending=False)
               .head(10))

# plotto avg_streams per track
top_artists['avg_streams'].plot(kind='bar', figsize=(10,6), color='green', edgecolor='black')
plt.title('Top 10 Artisti: Streams Medi per Traccia')
plt.ylabel('Streams Medi per Traccia')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# controllo tracce con lo stesso id
conteggi = ds['id'].value_counts()

duplicati = conteggi[conteggi > 1]

print(duplicati)
```

```python
# vedo gli ID assegnati a tracce diverse che abbiamo scoperto prima

duplicati = ds[ds.duplicated(subset=['id'], keep=False)]

conteggio_titoli = duplicati.groupby('id')['title'].nunique()
ids_con_titoli_diversi = conteggio_titoli[conteggio_titoli > 1].index

print(f"Number of ids that have different titles between their copies: {len(ids_con_titoli_diversi)}")

if len(ids_con_titoli_diversi) > 0:
    cols_to_show = ['id', 'title', 'primary_artist', 'album']
    display(ds[ds['id'].isin(ids_con_titoli_diversi)].sort_values('id')[cols_to_show])
```

```python
# Cerco canzoni duplicate (uso come chavi titolo e artista)
ds[ds.duplicated(subset=['title', 'primary_artist'], keep=False)]
```

```python
# Vedo se ci sono album con più di un primary artist assegnato
duplicati = ds.groupby('album').filter(lambda x: x['primary_artist'].nunique() > 1)

risultato = duplicati[['album', 'primary_artist', 'title']].drop_duplicates().sort_values('album')

pd.set_option('display.max_rows', None)

risultato.head(200)
```

```python
#Vedo se ci sono album con lo stesso id e li printo

cols_to_check = ['album_name', 'primary_artist']

grouped = ds.groupby('id_album')[cols_to_check].agg(lambda x: x.unique())

conflicts = grouped[grouped.apply(lambda row: any(len(x) > 1 for x in row), axis=1)]

print(conflicts)
```

```python
# Controllo se ogni artista ha almeno un valore BPM valido, che è un prerequisito per calcolare successivamente 
# le medie (per artista) e imputare i dati mancanti per le feature melodiche

# raggruppo per artista e conto i bpm validi
valid_bpm_counts = ds.groupby('primary_artist')['bpm'].count()

# vediamo se sono tutti sopra lo 0
result = (valid_bpm_counts > 0).all()

print(f"Do all artists have at least one track with a valid BPM? {result}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

# Crea una mappa dei colori personalizzata con un colore più arancione e nero
from matplotlib.colors import LinearSegmentedColormap

# Definisci i colori
colors = ["#00FF5987", "#0c0b00"]  # Nero e un arancione più arancione

# Crea la mappa dei colori
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors, N=2)

# Personalizza la dimensione della figura per la leggibilità
plt.figure(figsize=(8, 4))

# Trasponi il DataFrame per invertire gli assi
sns.heatmap(df.isnull().T, cbar=False, cmap=cmap)

# Aggiungi titolo e etichette per chiarezza
plt.xlabel("Features", fontsize=12)  # Features sull'asse x ora
# Rimuovi i numeri sull'asse delle x
plt.xticks([])
plt.yticks(fontsize=12)

plt.show()
```

```python
# Distribuzione per genere e nazionalità
df.groupby(['gender', 'nationality']).size().unstack(fill_value=0)
```

```python
# Età media artisti attivi
df['age'] = 2025 - pd.to_datetime(df['birth_date'], errors='coerce').dt.year
df.groupby('gender')['age'].agg(['mean', 'median', 'count'])
```

```python
# Merge tracks.json per conteggio canzoni/streams per artista
artist_stats = ds.groupby('id_artist').agg({
    'id': 'count', 
    'streams@1month': 'sum'
}).rename(columns={'id': 'n_tracks', 'streams@1month': 'total_streams'})

df_plot = df.merge(artist_stats, left_on='id_author', right_index=True, how='left').fillna(0)

# plotto la mappa
fig = px.scatter_mapbox(df_plot, 
                       lat='latitude', lon='longitude',
                       size='n_tracks', color='total_streams',
                       hover_name='name',
                       color_continuous_scale='viridis',
                       size_max=30,
                       zoom=6, center={'lat': 42.5, 'lon': 12.5},
                       mapbox_style='open-street-map',
                       title='Artisti Rap Italiani: Canzoni e Streams per Regione')

fig.show()
```
