```python
# 1. Calcoli i conteggi
conteggi = ds['id'].value_counts()

# 2. Filtri tenendo solo quelli maggiori di 1
duplicati = conteggi[conteggi > 1]

print(duplicati)
```

```python
# 1. Seleziona solo le righe che hanno ID duplicati
duplicati = ds[ds.duplicated(subset=['id'], keep=False)]

# 2. Raggruppa per 'id' e conta i titoli unici per ciascuno
# Se il risultato è > 1, significa che per quell'ID ci sono titoli diversi
conteggio_titoli = duplicati.groupby('id')['title'].nunique()
ids_con_titoli_diversi = conteggio_titoli[conteggio_titoli > 1].index

print(f"Numero di ID che hanno titoli diversi tra le loro copie: {len(ids_con_titoli_diversi)}")

# 3. Visualizza le righe "colpevoli"
if len(ids_con_titoli_diversi) > 0:
    cols_to_show = ['id', 'title', 'primary_artist', 'album']
    display(ds[ds['id'].isin(ids_con_titoli_diversi)].sort_values('id')[cols_to_show])
```

```python
# Duplicati basati su entrambe le colonne contemporaneamente
ds[ds.duplicated(subset=['title', 'primary_artist'], keep=False)]
```

```python
# Filtra album con artisti diversi
duplicati = ds.groupby('album').filter(lambda x: x['primary_artist'].nunique() > 1)

# Mostra solo album_name e primary_artist, rimuovi coppie duplicate
risultato = duplicati[['album', 'primary_artist', 'title']].drop_duplicates().sort_values('album')

# Mostra tutte le righe senza limiti
pd.set_option('display.max_rows', None)

risultato.head(200)
```
