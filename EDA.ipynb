```python
print(f"Righe totali ricaricate: {len(ds)}")

# 2. Ora cerchiamo i duplicati SU TUTTO il dataset
# keep=False ci mostra TUTTE le righe coinvolte nel duplicato
duplicati = ds[ds.duplicated(subset=['id'], keep=False)].sort_values('id')

print(f"Numero di righe duplicate trovate: {len(duplicati)}")

# 3. Vediamone un esempio concreto
if len(duplicati) > 0:
    print("\nEcco le prime righe duplicate a confronto:")
    # Mostriamo ID, Artista e Titolo per capire cosa cambia tra le copie
    display(duplicati[['id', 'primary_artist', 'title']].head(6))
```

```python
# 1. Isola le 144 righe che hanno l'ID duplicato
duplicati_id = ds[ds.duplicated(subset=['id'], keep=False)]

# 2. Cerca dentro questo gruppo le righe che sono identiche IN TUTTO (senza subset)
righe_identiche = duplicati_id[duplicati_id.duplicated(keep=False)]

print(f"Righe con ID uguale: {len(duplicati_id)}")
print(f"Righe completamente identiche (tutta la riga): {len(righe_identiche)}")

# 3. Se ce ne sono, mostriamole
if len(righe_identiche) > 0:
    display(righe_identiche.sort_values('id').head(4))
else:
    print("Nessuna riga è identica. Significa che cambiano sempre dei dati (es. titolo, album, features).")
```

```python
# 1. Seleziona solo le righe che hanno ID duplicati
duplicati = ds[ds.duplicated(subset=['id'], keep=False)]

# 2. Raggruppa per 'id' e conta i titoli unici per ciascuno
# Se il risultato è > 1, significa che per quell'ID ci sono titoli diversi
conteggio_titoli = duplicati.groupby('id')['title'].nunique()
ids_con_titoli_diversi = conteggio_titoli[conteggio_titoli > 1].index

print(f"Numero di ID che hanno titoli diversi tra le loro copie: {len(ids_con_titoli_diversi)}")

# 3. Visualizza le righe "colpevoli"
if len(ids_con_titoli_diversi) > 0:
    cols_to_show = ['id', 'title', 'primary_artist', 'album']
    display(ds[ds['id'].isin(ids_con_titoli_diversi)].sort_values('id')[cols_to_show])
else:
    print("Tutti gli ID duplicati hanno lo stesso titolo identico.")
```
